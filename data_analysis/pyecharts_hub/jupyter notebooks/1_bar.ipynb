```python
from pyecharts.charts import Bar # 导入Bar类
from pyecharts import options as opts # 导入options类
# 创建Bar实例
bar =Bar()
bar.add_xaxis(["衬衣","毛衣","领带","裤子","风衣","高跟鞋","袜子"])
bar.add_yaxis("商家A",[114,55,27,101,125,27,105])
bar.add_yaxis("商家B",[57,134,137,129,145,12,108])
bar.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例",subtitle="我是副标题"))
bar.render_notebook()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts


line = Line()
line.add_xaxis(["衬衫","毛衣","领带","裤子","风衣","高跟鞋","袜子"])
line.add_yaxis("商家A",[114,55,27,101,125,27,105])
line.add_yaxis("商家B",[57,134,137,129,145,12,108])
line.set_global_opts(title_opts=opts.TitleOpts(title="Line-基本示例",subtitle="我是副标题"))
line.render_notebook()
```

```python
from pyecharts.charts import Pie
from pyecharts import options as opts


pie = Pie()
pie.add("访问来源",[("直接访问",1250),("邮件营销",251),("联盟广告",135),("视频广告",1548),("搜索引擎",1208)])
pie.set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
pie.set_series_opts(label_opts=opts.LabelOpts(is_show=True))
pie.render_notebook()
```
