```python
import pandas as pd
```

```python
data = pd.read_csv("../data/学生表现因素.csv")
data.head()
```

```python
data.columns
```

```python
# 创建列名映射字典
column_mapping = {
    'Hours_Studied': '学习时长',
    'Attendance': '出勤率',
    'Parental_Involvement': '家长参与度',
    'Access_to_Resources': '资源获取',
    'Extracurricular_Activities': '课外活动',
    'Sleep_Hours': '睡眠时间',
    'Previous_Scores': '过往成绩',
    'Motivation_Level': '学习动机水平',
    'Internet_Access': '网络接入',
    'Tutoring_Sessions': '辅导课程',
    'Family_Income': '家庭收入',
    'Teacher_Quality': '教师质量',
    'School_Type': '学校类型',
    'Peer_Influence': '同伴影响',
    'Physical_Activity': '体育活动',
    'Learning_Disabilities': '学习障碍',
    'Parental_Education_Level': '家长教育水平',
    'Distance_from_Home': '家庭距离',
    'Gender': '性别',
    'Exam_Score': '考试成绩'
}

# 假设您的DataFrame变量名为df
data.rename(columns=column_mapping, inplace=True)

# 验证转换结果
print("转换后的列名：")
print(data.columns.tolist())
```

```python
data
```

```python
# 定义英文到中文的映射字典
translation_map = {
    'Low': '低',
    'Medium': '中',
    'High': '高',
    'No': '否',
    'Yes': '是',
    'Public': '公立',
    'Private': '私立',
    'Positive': '积极',
    'Negative': '消极',
    'Neutral': '中性',
    'High School': '高中',
    'College': '大学',
    'Postgraduate': '研究生',
    'Near': '近',
    'Moderate': '中等',
    'Far': '远',
    'Male': '男',
    'Female': '女'
}

# 优化后的翻译逻辑
for column in data.columns:
    # 获取列的唯一值
    unique_vals = data[column].unique()
    
    # 检查列数据类型
    if unique_vals.dtype.kind in 'iuf':  # 数值类型（整数、无符号整数、浮点）
        # 数值列直接跳过
        continue
    else:
        # 非数值列进行翻译
        # 只处理映射字典中存在的值
        data[column] = data[column].map(
            lambda x: translation_map.get(x, x)  # 存在映射则翻译，否则保留原值
        )
```

```python
data.to_csv("../data/学生表现因素_processed.csv",index=False)
```
